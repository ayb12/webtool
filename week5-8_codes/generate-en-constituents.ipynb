### Constituent Parsing for EN

```
# Load libraries and Constituency Parser for EN
import spacy
from benepar.spacy_plugin import BeneparComponent
import en_core_web_sm
nlp = en_core_web_sm.load()
nlp.add_pipe(BeneparComponent("benepar_en2"))  # constituency parser for EN
```

```
"""
Perform constituency parsing on TED Talk EN transcripts
See https://github.com/nikitakit/self-attentive-parser for the documentation of the `doc`.
"""
en_anno_to_constituents = dict()
for sent_id, annos in all_annotations('en').items():
    for anno in annos:
        doc = nlp(anno.tokenized_text)
        en_anno_to_constituents[sent_id] = doc
        break

dill.dump(en_anno_to_constituents, open("en_constituents.dill", 'wb'))
# dill.load(open("en_constituents.dill", 'rb'))
```
